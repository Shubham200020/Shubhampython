```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import seed
from random import randint
```

```python
df=pd.read_csv('drug-overdose-death-rates new.csv')
year=df['Year']
```

```python
name1="Cocaine overdose death rates (CDC WONDER)"
name2="Heroin overdose death rates (CDC WONDER)"
plot1=df[name1]
plot2=df[name2]
plt.plot(year,plot1,label="Cocaine")
plt.plot(year,plot2,label="Heroin")
plt.title('Cocaine and Heroin overdose death rates')
plt.show()
```

```python
plt.bar(year,plot1,label="Cocaine",color="green")
plt.bar(year,plot2,label="Heroin",color="yellow")
plt.title('Cocaine and Heroin overdose death rates')
plt.show()
```
